```python

import pandas as pd
import numpy as np

data = pd.read_csv('data/report14-18.txt', delimiter='\t')
data.head()
```

```python
# data['살인검거율'] = int(data['살인.1'])/int(data['살인']) * 100

data.dtypes # 데이터 타입 확인
```

```python
# 데이터에서 ,를 제거
for i in list(data.columns[2:]):
    data[i] = data[i].str.replace(',','')

data
```

```python
# 0번째 행을 삭제
data = data.drop(data.index[0])
data.head()
```

```python
# 타입을 실수로 바꾸어 주기
data = data.astype({'합계':'float','합계.1':'float','살인':'float','살인.1':'float',\
             '강도':'float','강도.1':'float', '강간강제추행':'float','강간강제추행.1':'float'\
             ,'절도':'float','절도.1':'float','폭력':'float','폭력.1':'float'})
```

```python
# 실수로 바꿈
data.dtypes # 데이터 타입 확인
```

```python
# 범죄 검거율
data['범죄검거율'] = round(data['합계.1']/data['합계']*100,1)
data['범죄검거율']
```

```python
# del data['살인검거율']
# data['살인'].groupby(data['자치구'])
# data['강도'].groupby(data['자치구'])

# 발생 건수만 뽑음
occur = data.loc[:,['자치구','살인', '강도', '강간강제추행', '절도','폭력','합계','합계.1']]
occur = occur.groupby(['자치구']).sum()
print(occur.shape)
occur
```

```python
# 범죄검거율 
occur['검거율'] = round(occur['합계.1']/occur['합계']*100,1)
occur.head()
```

```python
# column 이름 바꾸기
occur_arrest = occur.rename(columns={'합계':'발생합계', '합계.1':'검거합계'})
occur_arrest
```

```python
# txt파일로 저장
occur_arrest.to_csv('data/crime_in_seoul.csv', mode='w')
```
