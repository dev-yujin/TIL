
## 3. git 사용하기
### a. init
* `git init`명령어를 사용하면 폴더에 .git 폴더 생긴다

* master 글자가 생긴다
```c
i@DESKTOP-CS5B0E0 MINGW64 ~/Desktop/learn_git (master)
```

* `git init`을 취소하려면 `rm -rf .git`을 사용한다

### b. 명칭
* `master`가 나오면 repo가 되었다고 말한다.
    * repository

### c. status를 확인하고 바뀐 파일을 add하기

* `git status` 를 통해 현재 바뀐것이 있는지 확인하기

```
$ git status
On branch master

No commits yet

nothing to commit (create/copy files and use "git add" to track)

```

* `touch README.md` 실행 후 다시 `status`
    * 추가된 파일이 있지만 tracking하지 않겠다

```
$ git status

On branch master

No commits yet

Untracked files:
  (use "git add <file>..." to include in what will be committed)

        README.md

nothing added to commit but untracked files present (use "git add" to track)
```

* `git add README.md` 를 통해 파일을 대기열에 추가한다

### d. 파일을 수정해보기
* README.md 파일을 수정하고 저장한 후 `git status` 해보면

```
$ git status
On branch master

No commits yet

Changes to be committed:
  (use "git rm --cached <file>..." to unstage)

        new file:   README.md

Changes not staged for commit:
  (use "git add <file>..." to update what will be committed)
  (use "git checkout -- <file>..." to discard changes in working directory)

        modified:   README.md

Untracked files:
  (use "git add <file>..." to include in what will be committed)

        .idea/

```


### e. commit, log, unstage

* `git commit -m 'createREADME.md ' 으로 커밋할 수 있다.


* `git log`를 통해서 commit log를 볼 수 있다

```
$ git log
commit 5e6fedbed944b4f27c9d9f56d6a78276065be8c3 (HEAD -> master)
Author: dev-yujin <k6@gmail.com>
Date:   Thu Aug 20 15:05:12 2020 +0900

    <E3><85><8A><C3><A3><C2><84>createREADME.md

```

* `git reset HEAD README.md`을 통해 unstage 할 수 있다.
    * unstage 명령어는 `git add .` 한 후, `git status`를 확인해보면 알 수 있다.
    
* 폴더/파일을 삭제한 경우에도 '삭제한 사실'을 commit해야한다

### f. remote
* github는 git의 드라이브이다.
* github에 있는 repo에 local 폴더를 remote할 수 있다

* `git remote add origin https://github.com/dev-yujin/learn_git.git` 을 통해 github에 remote한다

```
$ git remote -v
origin  https://github.com/dev-yujin/learn_git.git (fetch)
origin  https://github.com/dev-yujin/learn_git.git (push)

```

* `git push origin master`를 통해 push한다

```
$ git push origin master
Counting objects: 14, done.
Delta compression using up to 8 threads.
Compressing objects: 100% (11/11), done.
Writing objects: 100% (14/14), 2.89 KiB | 1.45 MiB/s, done.
Total 14 (delta 1), reused 0 (delta 0)
remote: Resolving deltas: 100% (1/1), done.
To https://github.com/dev-yujin/learn_git.git
 * [new branch]      master -> master

```
